```python
from sympy import Symbol, Derivative, symbols
a= Symbol('a')
d= Symbol('d')
function = a*2 + d*2 + 10
partialderivative= Derivative(function, d)
dfd = partialderivative.doit()
partialderivative= Derivative(function, a)
dfa = partialderivative.doit()
a1 = -1.0 
d1 = 2.0
itr = 100  
learningrate = 0.1 
for i in range(0,itr):
    a = symbols('a')
    d = symbols('d')
    dfav = dfa.subs(a, a1) 
    dfav = round(dfav,2)
    dfdv = dfd.subs(d, d1) 
    dfdv = round(dfdv,2)
    da = (-1.0)*learningrate*dfav
    dd = (-1.0)*learningrate*dfdv
    a1 = a1 + da
    a1 = round(a1, 2) 
    d1 = d1 + dd
    d1 = round(d1, 2)
print(a1, d1)
print(f'minimum value obtained at a = {a1} ,d ={d1} for given function')
```
